```python
qualifying_nutrients = {nutr: info for nutr, info in nutrient_info.items() if info['type'] == 'qualifying'}
disqualifying_nutrients = {nutr: info for nutr, info in nutrient_info.items() if info['type'] == 'disqualifying'}
```

```python
def scale_for_nbc(df, nutrient_cols, scaling_factor=2000):
    """
    Scales nutrient intake values to a common energy level.
    
    Parameters:
      df             : DataFrame that contains the raw nutrient columns and 'energy_kcal_eaten'
      nutrient_info  : Dictionary with nutrient information (unit, target, type)
      scaling_factor : The energy value (in kcal) to which nutrient intakes are scaled (default=2000)
    """
    df = df.copy()
    for nutrient in nutrient_cols:
        conversion_factor = conversion_factors[nutrient_info[nutrient]['unit']]
        df[nutrient + "_scaled"] = (
            df[nutrient] * (scaling_factor / df["energy_kcal_eaten"]) * conversion_factor
        )
    return df



```

```python
def calculate_qi_ratios(row, nutrient_cols):
  
    qi_ratios = {}
    

    for nutrient in nutrient_cols:
        qi_ratios[nutrient + '_ratio'] = row[nutrient + '_scaled'] / qualifying_nutrients[nutrient]['target']

    qi_ratios['QI'] = np.sum([i for i in qi_ratios.values()]) / len(qualifying_nutrients)

    qi_ratios["combined_name"] = row["combined_name"]
    qi_ratios["food_group"] = row["food_group"]
    qi_ratios["energy_kcal_eaten"] = row["energy_kcal_eaten"]
    qi_ratios["eaten_quantity_in_gram"] = row["eaten_quantity_in_gram"]
    
    return qi_ratios
```

```python
def calculate_di_ratios(row, nutrient_cols):
  
    di_ratios = {}
    

    for nutrient in nutrient_cols:
        di_ratios[nutrient + '_ratio'] = row[nutrient + '_scaled'] / disqualifying_nutrients[nutrient]['target']

    di_ratios['DI'] = np.sum([i for i in di_ratios.values()]) / len(disqualifying_nutrients)

    di_ratios["combined_name"] = row["combined_name"]
    di_ratios["food_group"] = row["food_group"]
    di_ratios["energy_kcal_eaten"] = row["energy_kcal_eaten"]
    di_ratios["eaten_quantity_in_gram"] = row["eaten_quantity_in_gram"]
    
    return di_ratios
```

```python

df_food_scaled = scale_for_nbc(df_food_filtered.fillna(0), nutrient_info, scaling_factor=2000)
qi_ratios = df_food_scaled.apply(calculate_qi_ratios, axis=1, nutrient_cols=qualifying_nutrients)
qi_ratios = qi_ratios.apply(pd.Series)
di_ratios = df_food_scaled.apply(calculate_di_ratios, axis=1, nutrient_cols=disqualifying_nutrients)
di_ratios = di_ratios.apply(pd.Series)
```

```python
# QI distribution for a specific food group
def qi_distribution(food_group, df, nutrient_info, daily_energy_needs, plot=True):
    df_food_group = df[df['food_group'] == food_group].copy()
    
    qi_values = calculate_qi(df_food_group, nutrient_info, daily_energy_needs)
    
    if plot:
        plt.figure(figsize=(10, 6))
        # Plot histogram as density instead of frequency
        sns.histplot(qi_values, bins=20, kde=True, stat="density")
        plt.title(f'Distribution of QI values for : {food_group}')
        plt.xlabel('QI')
        plt.ylabel('Density')
        plt.show()
 
    stats = qi_values.describe()
    return stats
```

```python
# Boxplot distribution of QI value for each nutrient for a specific food group
def plot_qi_boxplots_by_nutrient(food_group, df, nutrient_info, daily_energy_needs):
 
    df_food_group = df[df['food_group'] == food_group].copy()
    
    # Consider only qualifying nutrients.
    qualifying_nutrients = {nutrient: info for nutrient, info in nutrient_info.items() if info['type'] == 'qualifying'}
    
    qi_data = {}
    
    for nutrient, info in qualifying_nutrients.items():
        target = info['target']

        # Skip if target is not positive.
        if target <= 0:
            continue  
        
        scaling_factor = daily_energy_needs / df_food_group['energy_kcal_eaten']
        
        qi_values = (df_food_group[nutrient] * scaling_factor) / target
        qi_data[nutrient] = qi_values
    
    # Create a DataFrame from the dictionnary qi_data
    qi_df = pd.DataFrame(qi_data)
    
    # Melt convert wide dataFrame into a long one
    qi_long = qi_df.melt(var_name='nutrient', value_name='qi_value')
    

    plt.figure(figsize=(14, 8))
    sns.boxplot(data=qi_long, x='nutrient', y='qi_value')
    plt.title(f'QI Distribution for each nutrient in : {food_group}')
    plt.xlabel('Nutrient')
    plt.ylabel('QI Value')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    return qi_long
```

```python
# Define a function to classify each milk product by its name.
def classify_milk(product_name):
    # Convert the name to lowercase for case-insensitive matching.
    name = product_name.lower() if isinstance(product_name, str) else ""
    
    # Check for keywords in order of specificity.
    if "almond" in name or "mandel" in name or "amande" in name:
        return "Almond Milk"
    elif "coconut" in name or "kokos" in name:
        return "Coconut Milk"
    elif "oat" in name or "hafer" in name:
        return "Oat Milk"
    elif "rice" in name or "reis" in name:
        return "Rice Milk"
    elif "soja" in name or "soy" in name:
        return "Soya Milk"
    elif "milch" in name or "milk" in name:
        # If none of the plant-based keywords match but the name still contains "milk" or "milch",
        # we assume it is a dairy milk.
        return "Dairy Milk"
    else:
        return "Other"

# Subset the milk data.
df_milk = df_food_filtered[df_food_filtered['food_group'] == 'milk'].copy()

# Create a new column 'milk_subgroup' by applying the classification function.
df_milk['milk_subgroup'] = df_milk['combined_name'].apply(classify_milk)

# Now you can see how many items are in each subgroup.
subgroup_counts = df_milk['milk_subgroup'].value_counts()
print("Milk subgroups counts:")
print(subgroup_counts)


```
